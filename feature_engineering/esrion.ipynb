# Data visualization/plots

```python
#plot number of missing values in each column using bar plot matplotlib
missing_values = trainingset.isnull().sum().sort_values(ascending=False)

#create barplot
plt.figure(figsize=(12,8))  # Optional: You can set the figure size
missing_values.plot(kind='bar')

#rotate the x-axis labels
plt.xticks(rotation=85)
plt.title('Proportion of missing values in each column')
plt.xlabel('Columns')
plt.ylabel('Proportion of missing values')
plt.show()
```

```python
#plot number of missing values in each column using bar plot matplotlib
missing_values = traintest.isnull().sum().sort_values(ascending=False)

#create barplot
plt.figure(figsize=(12,8))  # Optional: You can set the figure size
missing_values.plot(kind='bar')

#rotate the x-axis labels
plt.xticks(rotation=85)
plt.title('Proportion of missing values in each column')
plt.xlabel('Columns')
plt.ylabel('Proportion of missing values')
plt.show()
```
